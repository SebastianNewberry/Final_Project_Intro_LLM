## 🔧 Step 1: Set up the LLM via Ollama

```
from langchain_community.llms import Ollama

# Load the LLaMA model served locally via Ollama
llm = Ollama(model="llama3")  # You can use 'llama2', 'mistral', etc.
```

## 📄 Step 2: Load and process documents

```
from langchain.document_loaders import TextLoader
from langchain.text_splitter import RecursiveCharacterTextSplitter

# Load a sample .txt file (replace with your own)
loader = TextLoader("sample.txt")  # Make sure this file exists
documents = loader.load()

# Split into smaller chunks
text_splitter = RecursiveCharacterTextSplitter(chunk_size=500, chunk_overlap=100)
docs = text_splitter.split_documents(documents)
```

## 🧠 Step 3: Create Embeddings and Store in FAISS

```
from langchain.vectorstores import FAISS
from langchain.embeddings import OllamaEmbeddings

# Create embeddings using Ollama
embedding = OllamaEmbeddings(model="nomic-embed-text")  # or 'llama3' if it supports embeddings

# Embed and store documents in FAISS vector store
vectorstore = FAISS.from_documents(docs, embedding)
```

## 🧪 Step 4: RAG - Ask Questions with Retrieved Context

```
from langchain.chains import RetrievalQA

# Create RetrievalQA chain with LLaMA and FAISS retriever
qa_chain = RetrievalQA.from_chain_type(
    llm=llm,
    retriever=vectorstore.as_retriever(),
    return_source_documents=True
)

# Ask a question!
query = "What is this document about?"
result = qa_chain({"query": query})

print("Answer:", result["result"])
```

## 📌 Notes

- Make sure `sample.txt` exists in the same directory as the notebook.
- You can change `Ollama(model="...")` to use other models like `mistral`, `llama2`, `codellama`, etc., if installed.
- For larger-scale apps, consider using Chroma or other persistent vector stores instead of FAISS.
